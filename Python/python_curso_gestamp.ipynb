Escribir una función que reciba una muestra de números y devuelva los valores atípicos, es decir, los valores cuya puntuación típica sea mayor que 3 o menor que -3.
Nota: La puntuación típica de un valor se obtiene restando la media y dividiendo por la desviación típica de la muestra.

```python
lista = [4, 8, 15, 16, 23, 10000000]

import numpy as np

def outliers(numeros):
  media = np.mean(numeros)
  st = np.std(numeros)
  puntuaciones = list(map(lambda numero: (numero-media)/st, numeros))
  return list(filter(lambda numeros: abs(puntuacion)>=2, puntuaciones))


print(outliers(lista))
```

Escribir una función que reciba un DataFrame con el formato del ejercicio anterior, una lista de meses, y devuelva el balance (ventas - gastos) total en los meses indicados.

```python
def balance(df, lista_mes):
  df["balance"] = df.ventas- df.gastos
  filtered_rows = filter(lambda row: row['mes'] in lista_mes, df.to_dict('records'))
  balances = map(lambda row: row['balance'], filtered_rows)
  total_balance = sum(balances)
  return total_balance
balance(data,['enero','febrero'])

```
